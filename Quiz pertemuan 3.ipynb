### Melakukan visualisasi data kategorik

#### data kategorik yang divisualisasikan akan berupa Pie Chart dan Bar Chart

```python
age_counts = data1['Age'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(age_counts.index, age_counts.values)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

```python
Gender_counts = data1['Gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(Gender_counts.values, labels=Gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.show()
```

#### di atas merupakan hasil dari distribusi atau persebaran umur dan gender pada data yang saya punya

# Visualisasi data numerik

#### data numerik merupakan data yang diperoleh dari hasil perhitungan. data ini memiliki dua jenis, yaitu interval dan ratio. Visualisasi data numerik kali ini akan menggunakan Histogram dan Scatter plot.

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Histogram for Heart Rate
ax.hist(data1['Physical Activity Level'], bins= range(30, 80, 7), edgecolor='black')
ax.set_xlabel('Physical Activity Level')
ax.set_ylabel('Frequency')
ax.set_title('Heart Rate Distribution')

plt.show()
```

### data histogram di atas merupakan data berdasarkan kegiatan fisik dengan frekuensi

## selanjutnya akan memvisualisasikan hubungan antara physical activity level dengan heart rate

```python
plt.figure(figsize=(8, 6))

plt.scatter(data1['Physical Activity Level'], data1['Heart Rate'])
plt.xlabel('Physical Activity Level')
plt.ylabel('Heart Rate')
plt.title('Relationship between Physical Activity Level and Heart Rate')


plt.show()
```
